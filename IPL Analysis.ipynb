```python
def get_city_match_count(city):
    # Assuming 'data' is your IPL dataset
    mask = data['city'] == city  # Check if the 'city' column matches the specified city
    match_count = data[mask].shape[0]  # Count the number of rows where the condition is True
    return match_count
```

```python
def run_scored(batsman_name):
    mask=delivery[delivery["batsman"]==batsman_name]
    return mask.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False)
```

```python
run_scored("MS Dhoni")
```

```python
#Finding the most destructive batsman in 16 to 20 over with atleast 200 balls played

death_over=delivery[delivery["over"]>15]
```

```python
#couting no. of balls played by a batsman in 16-20 ovrs atleast 200 balls
all_batsman=death_over.groupby("batsman")["batsman_runs"].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()

```

```python
#runs scored by all these 43 batsman 
#balls played by these 43 batsman
final=delivery[delivery["batsman"].isin(batsman_list)]
```

```python
runs=final.groupby("batsman")["batsman_runs"].sum()
```

```python
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
sr=(runs/balls)*100
sr.sort_values(ascending=False)
```
