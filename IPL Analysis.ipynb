```python
def get_city_match_count(city):
    # Assuming 'data' is your IPL dataset
    mask = data['city'] == city  # Check if the 'city' column matches the specified city
    match_count = data[mask].shape[0]  # Count the number of rows where the condition is True
    return match_count
```

```python
#Group BY EDA IPL Analysis
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery
```

```python
batsman=delivery.groupby("batsman")
batsman
```

```python
len(batsman)
```

```python
#Batsman With Most Runs
batsman["batsman_runs"].sum().sort_values(ascending = False).head(10)
```

```python
#Most Foursa
mask=delivery["batsman_runs"]==4
new_delivery=delivery[mask]
new_delivery
```

```python
new_delivery.shape[0]
```

```python
four_batsmen=new_delivery.groupby("batsman")
four_batsmen["batsman_runs"].count().sort_values(ascending=False).head(10)
```

```python
delivery.head(3)
```

```python
#MAX runs scored by Virat Kohli against a teams


mask=delivery["batsman"]=="V Kohli"
mask
```

```python
kohli=delivery[mask]
kohli
```

```python
kohli_against=kohli.groupby("bowling_team").sum("batsman_runs")
kohli_against["batsman_runs"].sort_values(ascending=False).head(5)
```

```python
len(kohli.groupby("bowling_team"))
```

```python
def run_scored(batsman_name):
    mask=delivery[delivery["batsman"]==batsman_name]
    return mask.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False)
```

```python
run_scored("MS Dhoni")
```
